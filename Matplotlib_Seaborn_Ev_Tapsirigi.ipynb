### Tapşırıq 2: Tək Dəyişənli Analiz (Univariate Analysis)
Almazların qiymət (`price`) paylanmasını və sıxlığını təhlil edin. Bu növ analiz kəmiyyət dəyişəninin asimmetriyasını (skewness) anlamağa kömək edir.

**Tələblər:**
- `Seaborn` kitabxanasından `histplot` funksiyasını istifadə edin.
- Qrafikə Nüvə Sıxlığı Təxmini (KDE - Kernel Density Estimation) əyrisi əlavə edin.
- Zibil qutularının (bins) sayını 40 olaraq təyin edin.
- Oxlara ad (`xlabel`, `ylabel`) and qrafikə mərkəzləşdirilmiş başlıq (`title`) verin.

```python
plt.figure(figsize=(10, 6))

# Paylanma qrafikini qurun
sns.histplot(data=df,
             x='price',
             bins=40,
             kde=True)

plt.title("Almazların Qiymət Paylanması (Price Distribution)", fontsize=14, fontweight='bold')
plt.xlabel("Qiymət (USD)")
plt.ylabel("Tezlik")
plt.show()
```

### Tapşırıq 3: İki Dəyişənli Analiz (Bivariate Analysis) - Kateqorik və Kəmiyyət
Almazın kəsim keyfiyyətinin (`cut`) onun qiymətinə (`price`) təsirini araşdırın. Qutu qrafiki (Boxplot) median, kvartillər və anomaliyaları (outliers) eyni anda görmək üçün optimal vasitədir.

**Tələblər:**
- X oxunda kəsim keyfiyyətini (`cut`), Y oxunda isə qiyməti (`price`) yerləşdirin.
- `Seaborn` vasitəsilə `boxplot` qurun.
- Qrafikdə məlumatları daha aydın fərqləndirmək üçün uyğun bir rəng palitrası (`palette="Set2"`) təyin edin.

```python
plt.figure(figsize=(10, 6))

# Qutu qrafikini (boxplot) qurun
sns.boxplot(data=df,
            x='cut',
            y='price',
            palette='Set2')

plt.title("Kəsim keyfiyyətinə görə qiymət fərqlilikləri", fontsize=14)
plt.xlabel("Kəsim Keyfiyyəti (Cut)")
plt.ylabel("Qiymət (USD)")


plt.title("Kəsim keyfiyyətinə görə qiymət fərqlilikləri", fontsize=14)
plt.show()
```

### Tapşırıq 4: Çoxölçülü Analiz (Multivariate Analysis) və Korrelyasiya Matrisi
Məlumat dəstindəki kəmiyyət dəyişənləri (məsələn: `carat`, `depth`, `table`, `price`, `x`, `y`, `z`) arasındakı xətti asılılığı (Pearson korrelyasiyası) riyazi olaraq hesablayıb vizuallaşdırın. Multikollinearlıq problemini aşkarlamaq üçün bu addım kritikdir.

**Tələblər:**
1. Məlumat dəstindən yalnız ədədi sütunları ayıraraq korrelyasiya matrisini (`.corr()`) yaradın.
2. `Seaborn heatmap` funksiyasından istifadə edərək matrisi vizuallaşdırın.
3. Hər bir xananın üzərində korrelyasiya əmsalını rəqəmlə göstərin (`annot=True`).
4. Rəng xəritəsi kimi temperaturu ifadə edən palitra seçin (məs: `cmap="coolwarm"`).

```python
plt.figure(figsize=(10, 8))

# 1. Ədədi sütunların seçilməsi və korrelyasiyanın hesablanması
correlation_matrix = df.select_dtypes(include='number').corr()

# 2. İstilik xəritəsinin (heatmap) qurulması
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title("Kəmiyyət Dəyişənləri Arasında Pearson Korrelyasiya Matrisi", fontsize=14)
plt.show()

```

### Tapşırıq 5: Obyekt Yönümlü Matplotlib ilə Alt-Qrafiklər (Subplots)
Analitik hesabatlarda fərqli ölçüləri tək paneldə nümayiş etdirmək tələb olunur. `Matplotlib`-in obyekt yönümlü API-sindən (`fig, ax`) istifadə edərək 1 sətir və 2 sütundan ibarət kompleks fiqur yaradın.

**Tələblər:**
- **Sol panel (ax[0]):** `carat` və `price` arasındakı asılılığı göstərən səpilmə qrafiki (Scatter plot). Həddindən artıq məlumat nöqtəsinin qrafiki bağlamaması üçün şəffaflıq parametrini (`alpha=0.3`) təyin edin.
- **Sağ panel (ax[1]):** Almazın rəng kateqoriyalarına (`color`) görə ümumi tezliyini (sayını) göstərən sütunlu qrafik (Countplot - Seaborn ilə).

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# 1. Sol panel: Karat vs Qiymət səpilmə qrafiki
sns.scatterplot(data=df,
                x='carat',
                y='price',
                alpha=0.3,
                ax=axes[0])
axes[0].set_title("Karat və Qiymət Asılılığı (Scatter Plot)")
axes[0].set_xlabel("Karat (Carat)")
axes[0].set_ylabel("Qiymət (USD)")


# 2. Sağ panel: Rəng kateqoriyalarının tezliyi
sns.countplot(data=df,
              x='color',
              ax=axes[1])
axes[1].set_title("Rəng Kateqoriyalarına Görə Say (Countplot)")
axes[1].set_xlabel("Rəng Kodu")
axes[1].set_ylabel("Tezlik")

plt.tight_layout()
plt.show()
```
